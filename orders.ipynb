# Desafio: Estatistica

```python
import pandas as pd
import matplotlib.pyplot as plt
print(pd.__version__)
```

```python
# Dicionário de faturamento
dict_faturamento = {
    'data_ref': [
        '2023-01-01', 
        '2020-02-01', 
        '2021-03-01', 
        '2022-04-01', 
        '2023-05-01',
        '2023-06-01', 
        '2020-07-01', 
        '2021-08-01', 
        '2022-09-01', 
        '2023-10-01',
        '2022-11-01', 
        '2023-12-01',
        ],
    'valor': [
        400000, 
        890000, 
        760000, 
        430000, 
        920000,
        340000, 
        800000, 
        500000, 
        200000, 
        900000,
        570000, 
        995000,
        ]
}
```

```python
# Media das vendas
df_faturamento = pd.DataFrame.from_dict(dict_faturamento)
```

```python
df_faturamento
```

## Média

```python
media_vendas = round(df_faturamento.valor.mean(), 2)
print(media_vendas)
```

## Gráfico de Barra - Vendas

```python
# Gráfico de barras
chart = df_faturamento.plot.bar(title="Vendas por mês", x="data_ref", y="valor", figsize=(10, 5), color='blue', xlabel='Mês')
```

## Gráfico em Linha - Vendas

```python
# Gráfico de linha
chart = df_faturamento.plot.line(title="Vendas por mês", x="data_ref", y="valor", figsize=(10, 5), color='blue', xlabel='Mês')
```
